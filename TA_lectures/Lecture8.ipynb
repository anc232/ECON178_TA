### Working with Dummy Variables 


```r
data(Wage, package="ISLR")
head(Wage)
```

- We want to preduct log-wage using available information; 
- How should we pass dummy variables to the functions? 

```r
### Transform the matrix

new_X <- model.matrix(~. -1, data = Wage[, c(1:9)])
head(new_X)
```

What happened? 

- We transformed the matrix into a set of 0/1 dummy variables. Now every component is numeric! 

```r
### Run lasso and ridge regression 
library(glmnet)
my_lasso <- cv.glmnet(y = Wage$logwage, x = as.matrix(new_X), alpha = 1)
my_ridge <- cv.glmnet(y = Wage$logwage, x = as.matrix(new_X), alpha = 0)
plot(my_lasso)
plot(my_ridge)

```

```r
## Look at the coefficients
print(cbind(coef(my_lasso), coef(my_ridge)))
```

```r
## Interacting dummy variables 

X_with_interactions <- cbind(new_X, new_X[, 3:7]*new_X[,8], new_X[, 3:7]*new_X[,9], new_X[, 3:7]*new_X[,10])
head(X_with_interactions)
```

```r
### Short-cuts: looking at MSPE 
## Return the optimal MSPE
c(min(my_lasso$cvm), min(my_ridge$cvm))
```

```r
## Compare for example to stepwise : you should run cross-validation 

## You should do stepwise 

n <- length(Wage$logwage)
k <- 5
ii <- sample(rep(1:k, length= n))
pr.stepwise_backward <- pr.stepwise_forward <- rep(NA, n)
library(MASS)
for (j in 1:k){
  hold <- (ii == j)
  train <- (ii != j)
  ## Stepwise 
  full <- lm(Wage$logwage[train] ~ ., data= as.data.frame(new_X[train,]))
  null <- lm(Wage$logwage[train] ~ 1, data= as.data.frame(new_X[train,]))
  a <- stepAIC(null, scope=list(lower=null, upper=full), direction='forward')
  # backward stepwise - AIC
  b <- stepAIC(full, scope=list(lower=null, upper=full), direction='backward')
  pr.stepwise_backward[hold] <- predict(b, newdata= as.data.frame(new_X[hold,]))
  pr.stepwise_forward[hold] <- predict(a, newdata= as.data.frame(new_X[hold,]))
  }
mean((pr.stepwise_backward - Wage$logwage)**2, (pr.stepwise_forward - Wage$logwage)**2)
```

```r
## Use lasso to select spline methods 

## Create new matrices of splines 
kn <- as.numeric( quantile(Boston$dis, (1:4)/5) )

# prepare the matrix of covariates
x <- matrix(0, dim(Boston)[1], length(kn)+2)
for(j in 1:length(kn)) {
  x[,j] <- pmax(Boston$dis-kn[j], 0)^3 ## <<<---- Note: here you put the degree
}
x[, length(kn)+1] <- Boston$dis
x[, length(kn)+2] <- Boston$dis^2   


## Repeat with a second variable

kn <- as.numeric( quantile(Boston$ptratio, (1:4)/5) )

# prepare the matrix of covariates
x2 <- matrix(0, dim(Boston)[1], length(kn)+2)
for(j in 1:length(kn)) {
  x2[,j] <- pmax(Boston$ptratio-kn[j], 0)^3 ## <<<---- Note: here you put the degree
}
x2[, length(kn)+1] <- Boston$ptratio
x2[, length(kn)+2] <- Boston$ptratio^2   

```

```r
## New matrix
XX <- cbind(x, x2)
## Run lasso

my_lasso <- cv.glmnet(y = Boston$nox, x = XX)
plot(my_lasso)
coef(my_lasso)
```
