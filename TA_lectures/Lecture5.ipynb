### Step Wise: Recap

We can perform: 

- Backward Step Wise
- Forward Step Wise

Can we do it in a few lines of code? 

```r
#install.packages('MASS')
library(MASS)

## Load the data 
Credit <- read.csv("https://raw.githubusercontent.com/dviviano/ECON178_TA/master/data/Credit.csv")
```

```r
## Define a model with all and none of the variables
## We remove the first variable from Credit
full <- lm(Balance ~ ., data=Credit[,-1])
null <- lm(Balance ~ 1, data=Credit[,-1])
```

```r
# forward stepwise - AIC
a <- stepAIC(null, scope=list(lower=null, upper=full), direction='forward')

# backward stepwise - AIC
b <- stepAIC(full, scope=list(lower=null, upper=full), direction='backward')
```

```r
## Look at the coefficients

coef(a)
coef(b)
```

```r
## Look at AIC
extractAIC(a)
extractAIC(b)
```

```r
## The two methods lead to the same result. 
```

But an alternative method may be...
$$
\beta^{Lasso} = \underset{\beta}{\operatorname{argmin}}  \sum_{i=1}^n (Y_i - X_i\beta)^2 + \lambda \sum_{j=1}^p |\beta_j| 
$$


<br />


**Interpretation and differences with Ridge Regression**

- Lasso shrink some coefficients to zero after a certain treshold for lambda (features selection).
- Ridge never shrinks coefficients completely to zero.

```r
library(glmnet)

## Define the variables 
Credit <- read.csv("https://raw.githubusercontent.com/dviviano/ECON178_TA/master/data/Credit.csv")
y <- Credit$Balance 
X <- Credit[, c(2:11)] ## Save your covariates
head(X)
```

```r
### We need to force the factor to be numeric
## Force the matrix X to be numeric 
X <- model.matrix(~. ,data=X)[,-1]
head(X)
```

```r
### Let's Perform Lasso
## Choose lambda via cross-validation
lambda_seq = exp(seq(from = -4, to = 4, length = 50))
lasso_reg <- cv.glmnet(y = y, x = as.matrix(X), alpha = 1, lambda = lambda_seq)
ridge_reg <- cv.glmnet(y = y, x = as.matrix(X), alpha = 0, lambda = lambda_seq)
```

```r
plot(lasso_reg)
```

```r
plot(ridge_reg)
```

```r
### Look at the coefficients
cbind(coef(lasso_reg), coef(ridge_reg))
```

```r
## We look at the coefficients as a function of the lambda parameter
## We run a regression after removing the StudentYes variable (arbitrary choice)
lasso_reg <- glmnet(y = y, x = as.matrix(X[,-8]), alpha = 1)
ridge_reg <- glmnet(y = y, x = as.matrix(X[,-8]), alpha = 0)

plot(lasso_reg)
```

### Exercise 

- Perform Cross validation to compare Forward and Backward Step wise with Lasso and Ridge regression. Which model do you choose? 

```r
### Solution 


n <- length(y)
k <- 10
ii <- sample(rep(1:k, length= n))
pr.stepwise_backward <- pr.stepwise_forward <- pr.lasso <- pr.ridge <- rep(NA, length(y))

for (j in 1:k){
    
  hold <- (ii == j)
  train <- (ii != j)
  ## Stepwise 
  full <- lm(Balance ~ ., data=Credit[train,-1])
  null <- lm(Balance ~ 1, data=Credit[train,-1])
  a <- stepAIC(null, scope=list(lower=null, upper=full), direction='forward')
  # backward stepwise - AIC
  b <- stepAIC(full, scope=list(lower=null, upper=full), direction='backward')
  pr.stepwise_backward[hold] <- predict(b, newdata=Credit[hold,])
  pr.stepwise_forward[hold] <- predict(a, newdata=Credit[hold,])
  
  ## Do with lass (we use X and y defined above)
  xx.tr <- X[train,]
  y.tr <-  y[train]
  xx.te <- X[hold,]
  ridge.cv <- cv.glmnet(x=as.matrix(xx.tr), y=y.tr, nfolds=k, alpha=0)
  lasso.cv <- cv.glmnet(x=as.matrix(xx.tr), y=y.tr, nfolds=k, alpha=1)
  pr.lasso[hold] <- predict(ridge.cv, newx=xx.te)
  pr.ridge[hold] <- predict(lasso.cv, newx=xx.te)
  }

mspe_step_backward <- mean((pr.stepwise_backward-y)^2)
mspe_step_forward <- mean((pr.stepwise_forward-y)^2)
mspe.Lasso <- mean((pr.lasso-y)^2)
mspe.ridge <- mean((pr.ridge-y)^2)


```

```r
print(c(mspe_step_backward, mspe_step_forward, mspe.Lasso, mspe.ridge))
```
